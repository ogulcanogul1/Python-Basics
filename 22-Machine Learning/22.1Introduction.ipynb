# 🤖 Yapay Zeka (AI), Makine Öğrenmesi (ML), Derin Öğrenme (DL) ve Veri Bilimcisi (Data Scientist)

---

## 🔹 1. Yapay Zeka (Artificial Intelligence - AI)  
**Tanım:**  
Yapay Zeka, bilgisayarların ve makinelerin **insan zekasını taklit etmesini** sağlayan geniş bir alandır.  
Amaç; düşünme, öğrenme, karar verme, problem çözme gibi insan yeteneklerini bilgisayarlara kazandırmaktır.  

**Özellikleri:**
- İnsan gibi karar alabilme.
- Deneyimlerden öğrenme.
- Problemleri çözme.
- Doğal dil işleme (NLP), görüntü işleme, robotik gibi alt dalları içerir.

**Örnekler:**  
- Siri, Alexa gibi sanal asistanlar.  
- Otonom araçlar.  
- Satranç oynayan yapay zekalar.  

---

## 🔹 2. Makine Öğrenmesi (Machine Learning - ML)  
**Tanım:**  
AI’ın bir alt alanıdır.  
Makine öğrenmesi, bilgisayarların **verilerden öğrenmesini** ve **tecrübe kazandıkça performansını geliştirmesini** sağlar.  
Yani, açıkça programlanmadan **verilerden kurallar çıkarır**.

**Türleri:**
1. **Denetimli Öğrenme (Supervised Learning):**  
   - Etiketli verilerle çalışır (girdi + doğru çıktı).  
   - Örn: Ev fiyatı tahmini, e-posta spam filtresi.  

2. **Denetimsiz Öğrenme (Unsupervised Learning):**  
   - Etiketsiz verilerle çalışır, verilerdeki gizli yapıları bulur.  
   - Örn: Müşteri segmentasyonu, kümeleme.  

3. **Pekiştirmeli Öğrenme (Reinforcement Learning):**  
   - Deneme-yanılma ile öğrenir.  
   - Örn: Robotun yürümesi, oyun oynayan yapay zekalar.

**Örnek:**  
Netflix’in kullanıcıya film önermesi, spam e-posta filtresi.  

---

## 🔹 3. Derin Öğrenme (Deep Learning - DL)  
**Tanım:**  
Makine öğrenmesinin bir alt alanıdır.  
**Yapay sinir ağları** üzerine kuruludur ve özellikle çok büyük veri setlerinde başarılıdır.  

**Özellikleri:**  
- İnsan beynini taklit eden çok katmanlı yapay sinir ağları kullanır.  
- Görüntü işleme, ses tanıma, doğal dil işleme gibi **karmaşık verilerde** etkilidir.  
- Yüksek işlem gücü (GPU/TPU) gerektirir.  

**Örnekler:**  
- Google Translate’in sesli çeviri yapması.  
- Otonom araçların yol algılaması.  
- Yüz tanıma sistemleri.  

---

## 🔹 4. Veri Bilimcisi (Data Scientist)  
**Tanım:**  
Veri bilimcisi, **verilerden anlam çıkaran**, bu verileri analiz eden, modeller kuran ve iş dünyasına değerli içgörüler sunan kişidir.  

**Görevleri:**
- Veri toplama ve temizleme.  
- Veri görselleştirme.  
- İstatistiksel analiz.  
- Makine öğrenmesi ve derin öğrenme modelleri geliştirme.  
- İş kararlarını destekleyecek tahminler ve öneriler üretme.  

**Kullandığı Araçlar ve Teknolojiler:**  
- Programlama: Python, R, SQL.  
- Kütüphaneler: Pandas, NumPy, Scikit-learn, TensorFlow, PyTorch.  
- Görselleştirme: Matplotlib, Seaborn, PowerBI, Tableau.  
- Büyük veri: Hadoop, Spark.

# 📘 Makine Öğrenmesi (Machine Learning) Öğrenme Teknikleri

Makine Öğrenmesi (ML), bilgisayarların verilerden öğrenmesini ve tahminler yapmasını sağlar.  
Üç temel öğrenme tekniği vardır: **Denetimli Öğrenme (Supervised Learning), Denetimsiz Öğrenme (Unsupervised Learning), Pekiştirmeli Öğrenme (Reinforcement Learning).**

## 🔹 1. Denetimli Öğrenme (Supervised Learning)

**Tanım:**  
Denetimli öğrenmede elimizde **etiketli veri** bulunur.  
Model, giriş (X) ve doğru çıkış (y) verilerinden öğrenir. Daha sonra yeni veriler geldiğinde doğru tahmini yapmaya çalışır.  

**Özellikleri:**  
- Girdi + çıktı verisi bilinir.  
- Amaç: Tahmin yapmak (prediction).  
- Model hatalarını görerek kendini düzeltir.  

**Alt Türleri:**  
1. **Regresyon (Regression):**  
   - Sürekli değer tahmin eder.  
   - Örn: Ev fiyatı tahmini, sıcaklık tahmini.  
   
2. **Sınıflandırma (Classification):**  
   - Veriyi sınıflara ayırır (0-1, k kategorisi).  
   - Örn: E-posta spam mi değil mi, müşteri churn tahmini.  

**Örnek Algoritmalar:**  
- Linear Regression, Logistic Regression  
- Decision Trees, Random Forest  
- Support Vector Machines (SVM)  
- Neural Networks  

---

## 🔹 2. Denetimsiz Öğrenme (Unsupervised Learning)

**Tanım:**  
Denetimsiz öğrenmede elimizde **etiketli çıktı yoktur**.  
Model, sadece girdilerden (X) yararlanarak veriler arasındaki gizli yapıyı veya kalıpları keşfetmeye çalışır.  

**Özellikleri:**  
- Çıktı bilgisi yoktur.  
- Amaç: Veriyi gruplama, boyut azaltma, gizli ilişkiler bulma.  
- Daha çok keşifsel veri analizi (EDA) için kullanılır.  

**Alt Türleri:**  
1. **Kümeleme (Clustering):**  
   - Verileri benzerliklerine göre gruplar.  
   - Örn: Müşteri segmentasyonu (sadık müşteri, yeni müşteri vb.)  
   - Algoritmalar: K-Means, Hierarchical Clustering, DBSCAN  

2. **Boyut İndirgeme (Dimensionality Reduction):**  
   - Çok sayıda özelliği daha az boyuta indirerek anlamlı hale getirir.  
   - Örn: Görüntü sıkıştırma, PCA (Principal Component Analysis).  

**Örnek Algoritmalar:**  
- K-Means  
- PCA  
- t-SNE  
- Autoencoders  

---

## 🔹 3. Pekiştirmeli Öğrenme (Reinforcement Learning - RL)

**Tanım:**  
Pekiştirmeli öğrenmede **bir ajan (agent)** vardır ve bu ajan, bir ortam (environment) içinde hareket eder.  
Ajan, her adımda **ödül (reward)** veya **ceza (penalty)** alarak **deneme-yanılma yoluyla öğrenir**.  

**Özellikleri:**  
- Amaç: En iyi stratejiyi (policy) öğrenmek.  
- Ajan, uzun vadeli ödülü maksimize etmeye çalışır.  
- Denetimsiz veya denetimli gibi sabit veri yoktur, ajan çevreyle etkileşim halinde öğrenir.  

**Örnekler:**  
- Satranç / Go oynayan yapay zekalar (AlphaGo).  
- Otonom araçların trafik içinde karar vermesi.  
- Robotların yürüme öğrenmesi.  
- Reklamlarda en uygun kampanyayı seçen sistemler.  

**Temel Kavramlar:**  
- **Agent (Ajan):** Öğrenen model.  
- **Environment (Çevre):** Ajanın bulunduğu ortam.  
- **State (Durum):** Ajanın içinde bulunduğu anlık bilgi.  
- **Action (Eylem):** Ajanın yaptığı hareket.  
- **Reward (Ödül):** Ajanın yaptığı harekete verilen geri bildirim.  

**Örnek Algoritmalar:**  
- Q-Learning  
- Deep Q Networks (DQN)  
- Policy Gradient Methods

# 🌳 Karar Ağaçları ve Ansambl Yöntemler

Makine öğrenmesinde bazı algoritmalar **hem sınıflandırma (classification)** hem de **regresyon (regression)** problemlerinde kullanılabilir.  
Aradaki fark:  
- **Classification:** Sonuç kategorik → Örn: "evet/hayır", "hastalık var/yok".  
- **Regression:** Sonuç sürekli bir sayı → Örn: Ev fiyatı, sıcaklık, maaş tahmini.  

---

Random Forest,
AdaBoost,
XgBoost,

1) Instance-Based Learning (Örnek Tabanlı Öğrenme)

Tanım: Model, verileri öğrenirken aslında bütün örnekleri saklar ve yeni gelen veriler için bu örneklere bakarak karar verir.

Yani "ezberci" bir öğrenme yöntemidir.

Tahmin yaparken: Yeni bir veri geldiğinde en yakın örneklere bakılır → "Benzer veriler böyle sınıflandırılmış, o zaman bu da öyledir."

Örnekler:

KNN (K-Nearest Neighbors) → Yeni veriye en yakın komşulara bakıp sınıf veya değer tahmini yapar.

Case-Based Reasoning sistemleri.

✅ Avantaj:

Basit, genellikle küçük veri setlerinde iyi çalışır.

Non-linear (karmaşık) dağılımları yakalayabilir.

❌ Dezavantaj:

Bütün veriyi sakladığı için hafıza maliyeti yüksektir.

Tahmin yapmak yavaştır (çünkü her seferinde tüm veriyle karşılaştırma yapılır).

Gürültüye (noise) duyarlıdır.

--- 


2) Model-Based Learning (Model Tabanlı Öğrenme)

Tanım: Model, verilerden genel bir fonksiyon/hipotez öğrenmeye çalışır.

Yani "genelleme" yapar, ezberlemez.

Tahmin yaparken: Veriden çıkardığı matematiksel modeli/fonksiyonu kullanır → veriler saklanmaz, sadece model parametreleri saklanır.

Örnekler:

Lineer Regresyon (Linear Regression)

Decision Trees

Neural Networks

✅ Avantaj:

Öğrendikten sonra tahmin çok hızlıdır (sadece fonksiyonu uygular).

Büyük veri setlerinde daha verimli.

Gürültüye karşı daha dayanıklı olabilir.

❌ Dezavantaj:

Uygun bir model seçmek zor olabilir.

Model yanlış seçilirse underfitting/overfitting problemleri olur.
