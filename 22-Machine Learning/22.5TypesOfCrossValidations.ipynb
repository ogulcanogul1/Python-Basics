1️⃣ K-Fold Cross Validation

Mantığı:

Veri setini eşit parçalara (fold) böleriz. Diyelim veri setimiz 100 örnekten oluşuyor ve k=5 seçtik → her fold 20 örnek olacak.

Modeli 5 kez eğitiriz:

fold test, kalan 4 fold eğitim

fold test, kalan 4 fold eğitim
…

fold test, kalan 4 fold eğitim

Her seferinde farklı 20 örnek test olur, sonra tüm test sonuçlarının ortalaması alınır.

Avantaj:

Tüm veri hem eğitim hem test için kullanılır → veri israfı yok

Daha güvenilir performans tahmini sağlar

- Fold 1:  [T] E E E E
- Fold 2:  E [T] E E E
- Fold 3:  E E [T] E E
- Fold 4:  E E E [T] E
- Fold 5:  E E E E [T]
- T → Test, E → Eğitim

2️⃣ Stratified K-Fold

Mantığı:

K-Fold’un özel bir hali. Özellikle sınıflar dengesiz olduğunda kullanılır.

Örneğin: %90 sınıf A, %10 sınıf B. Normal K-Fold’da bazı fold’lar sadece A sınıfı olabilir → yanlış performans ölçümü.

Stratified K-Fold, her fold’da sınıf dağılımını orijinal veri setiyle aynı tutar.

Görsel olarak hayal et:

Orijinal veri: 90A, 10B

Fold 1: 18A, 2B  ← sınıf oranı korunuyor

Fold 2: 18A, 2B
…

3️⃣ Leave-One-Out (LOOCV)

Mantığı:

Veri setinde 1 örnek test, geri kalan tüm veri eğitim.

Toplam n örnek varsa n kez model eğitilir.

Küçük veri setlerinde maksimum veri kullanımı sağlar.

Avantaj:

Küçük veri setlerinde çok iyi tahmin sağlar
Dezavantaj:

Büyük veri setlerinde çok yavaş

Görsel:

Örnek sayısı = 5

Iterasyon 1: [T] E E E E

Iterasyon 2: E [T] E E E

Iterasyon 3: E E [T] E E

Iterasyon 4: E E E [T] E

Iterasyon 5: E E E E [T]


Burada her örnek test rolünü tek başına alır.

4️⃣ Leave-P-Out

Mantığı:

LOOCV’nin genellemesi.

Test setine her seferinde p örnek koyarsın, geri kalanı eğitim için kullanılır.

Örnek: Veri seti 5, p=2 → tüm 2’li kombinasyonlar test olur.

Avantaj: Küçük veri setlerinde farklı kombinasyonları sınamak mümkün.
Dezavantaj: Kombinasyon sayısı çok hızlı artar → büyük veri setlerinde hesaplama imkansız olur.

5️⃣ Repeated K-Fold

Mantığı:

K-Fold CV’yi birden fazla kez farklı random bölmelerle uygular.

Hedef: Tek bir K-Fold’un rastgele bölünme hatasını azaltmak.

Örnek: K-Fold’u 5 kez tekrar et, her seferinde farklı foldlar rastgele seçilir → sonuçların ortalaması alınır.

Avantaj: Daha güvenilir ve stabil performans ölçümü verir.

6️⃣ Time Series Cross Validation (Forward Chaining)

Mantığı:

Zaman serisi verilerinde, geçmiş verilerle geleceği tahmin etmek gerekir → random fold kullanılamaz.

Eğitim seti her iterasyonda büyür, test seti ise gelecekteki veriler olur.

Örnek: 

Iterasyon 1: Eğitim = [1], Test = [2]

Iterasyon 2: Eğitim = [1,2], Test = [3]

Iterasyon 3: Eğitim = [1,2,3], Test = [4]

🔹 1️⃣ K-Fold Cross Validation

Mantık:

Veri setini k eşit parçaya bölüyoruz (örneğin k=5).

Her seferinde bir fold test, kalan k-1 fold eğitim olur.

Toplam k kez eğitim/test yapılır.

Örnek:

Veri seti = 100 örnek

k = 5 → her fold 20 örnek

Fold 1: Test=1-20, Eğitim=21-100
Fold 2: Test=21-40, Eğitim=1-20 + 41-100
Fold 3: Test=41-60, Eğitim=1-40 + 61-100
Fold 4: Test=61-80, Eğitim=1-60 + 81-100
Fold 5: Test=81-100, Eğitim=1-80


Özellik:

Test seti her seferinde 20 örnek içeriyor.

Toplam eğitim/test çalıştırma sayısı = k = 5


---

🔹 2️⃣ Leave-One-Out Cross Validation (LOOCV)

Mantık:

Her seferinde sadece 1 örnek test, geri kalan tüm veri eğitim.

Toplam n örnek varsa n kez model eğitilir ve test edilir.

Örnek:

Veri seti = 100 örnek

LOOCV → test seti her seferinde 1 örnek, eğitim = 99 örnek

Iterasyon 1: Test=1, Eğitim=2-100
Iterasyon 2: Test=2, Eğitim=1 + 3-100
Iterasyon 3: Test=3, Eğitim=1-2 + 4-100
...
Iterasyon 100: Test=100, Eğitim=1-99


Özellik:

Test seti çok küçük (1 örnek), eğitim n-1 örnek

Toplam eğitim/test çalıştırma sayısı = n = 100 → çok daha fazla eğitim çalıştırılır


Özet Mantık:

K-Fold: Test seti makul boyutta, eğitim hızlı → büyük veri setleri için ideal

LOOCV: Her örnek tek tek test → küçük veri setlerinde maksimum veri kullanımı sağlar, ama büyük veri setlerinde hesaplama çok yavaş
