```python
df.fillna(0) ## missing valueları 0 ile doldurur.
df.head(2)
```

```python
## fillNa daha gerçekçi kullanımı
df['Sales_fillNA'] = df['Sales'].fillna(df['Sales'].mean())
df.head(15)
```

```python
## Renaming Columns

df = df.rename(columns={"Sale Date" : "Sales Date"})
df.head()
```

```python
## change datatypes
df['New_Sales'] = df['Sales'].fillna(df['Sales'].mean()).astype(int)
df.head()
```

```python
df['New_Value'] = df['Value'].apply(lambda x : x*2)
df.head()
```

```python
## Group By
group_mean = df.groupby('Product')['Sales'].mean()
print(group_mean)
```

```python
group_max = df.groupby(['Product','Region'])['Value'].max()
group_max
```

```python
## aggregate multiple functions
group_agg =  df.groupby('Region')['Value'].agg(['sum','mean','count'])
group_agg
```

#### pd.merge(df1, df2, on="Key", how="inner")

- Sadece iki tabloda da ortak olan satırlar gelir.

- Kesişim mantığı.

- Eğer bir Key sadece df1’de varsa ama df2’de yoksa, o satır alınmaz.

------------------------------------------------------------------------------
#### pd.merge(df1, df2, on="Key", how="outer")

- Her iki tablodaki tüm satırlar gelir.

- Ortak olmayanlara NaN konur.

------------------------------------------------------------------------------
#### 3. left  pd.merge(df1, df2, on="Key", how="left")

- (Sol JOIN – SQL LEFT JOIN)

- df1’deki tüm satırlar gelir.

- Eğer eşleşme yoksa df2 tarafı NaN olur.

------------------------------------------------------------------------------
#### 4. right pd.merge(df1, df2, on="Key", how="right")

- (Sağ JOIN – SQL RIGHT JOIN)

- df2’deki tüm satırlar gelir.

- Eğer eşleşme yoksa df1 tarafı NaN olur.



```python
## Merging and Joining Dataframes

df1 = pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','Y'],'Value2':[10,20,30]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='Key',how='inner') ## ortak değerler alınır
```

```python
pd.merge(df1,df2,on='Key',how='outer')
```

```python
pd.merge(df1,df2,on='Key',how='left')
```

```python
pd.merge(df1,df2,on='Key',how='right')
```
