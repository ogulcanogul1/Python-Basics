#### Değişken Standardizasyonu (Veri Standardizasyonu)

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])

df = pd.DataFrame({
    "V1":V1,
    "V2":V2,
    "V3":V3
})
df = df.astype(float)
df
```

#### Standardizasyon

```python
from sklearn import preprocessing
preprocessing.scale(df)
```

```python
df.mean()
```

#### Normalizasyon

```python
preprocessing.normalize(df)
```

#### Min-Max Dönüşümü

```python
preprocessing.minmax_scale(df,feature_range=(10,20)) ## 10-20 arasına dönüştürdü
scaler = preprocessing.MinMaxScaler(feature_range=(15,50))
scaler.fit_transform(df)
```

---

Standardizasyon (Z-score) : Ortalama 0, std 1

Normalizasyon : 0–1

Min Max : (veya belirlenen başka aralık)
