# Bernoulli Dağılımı (Discrete, PMF)

## Tanım
Bernoulli dağılımı, **tek bir denemenin sonucunu** modelleyen en basit olasılık dağılımıdır.  
Deneme iki sonuçtan birini verebilir:  
- Başarı (genellikle 1 ile gösterilir)  
- Başarısızlık (genellikle 0 ile gösterilir)  

Parametre olarak sadece **başarı olasılığı $p$** kullanılır. Başarısızlık olasılığı ise $1-p$ olur.  

## PMF (Probability Mass Function)
Bir Bernoulli rastgele değişkeni $X$ için PMF şu şekilde tanımlanır:

$$
P(X = x) =
\begin{cases}
p & x = 1 \\
1-p & x = 0
\end{cases}
$$

- $P(X=1)$ → Başarı olasılığı  
- $P(X=0)$ → Başarısızlık olasılığı  

---

## Özellikler
- **Kesikli dağılımdır** (discrete).  
- Tek parametre vardır: $p$ (başarı olasılığı).  
- Ortalama (expectation): $E[X] = p$  
- Varyans: $\text{Var}(X) = p(1-p)$  
- Tersine simetrik: Eğer $p=0.5$ ise dağılım simetriktir.

---

## Örnek 1: Yazı-Tura Oyunu
- Bir madeni para atılıyor.  
- Başarı = "Yazı gelmesi" (1)  
- Başarısızlık = "Tura gelmesi" (0)  
- Madeni para adil ise $p = 0.5$  

$$
P(X=1) = 0.5, \quad P(X=0) = 0.5
$$

Yani tek bir atışta yazı gelme olasılığı %50, tura gelme olasılığı %50’dir.

---

## Örnek 2: Bir Öğrencinin Sınavı Geçme Durumu
- Bir öğrenci bir sınavı geçebilir veya geçemeyebilir.  
- Başarı = sınavı geçmek (1)  
- Başarısızlık = sınavı geçememek (0)  
- Öğrencinin sınavı geçme olasılığı $p = 0.8$  

$$
P(X=1) = 0.8, \quad P(X=0) = 0.2
$$

Bu durumda öğrenci sınavı geçme olasılığı %80, geçememe olasılığı %20’dir.
